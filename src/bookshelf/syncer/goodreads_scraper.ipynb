```python
import requests
from bs4 import BeautifulSoup
```

```python
def get_html(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
    }
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        return response.text
    raise Exception(f"Error get_html status_code={response.status_code} response.text={response.text}")
def get_profile_html():
    html = get_html("https://www.goodreads.com/user/show/115130270-ben-rombaut")
    return html

def get_currently_reading_books():
    profile_html = get_profile_html()
    soup = BeautifulSoup(profile_html, 'html.parser')
    currently_reading_parent_div = soup.find('div', {'id': 'currentlyReadingReviews'})
    result = currently_reading_parent_div
    return result

get_currently_reading_books()
```

```python


def get_user_books(username):
    url = f"https://www.goodreads.com/user/show/{username}"
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
    }
    response = requests.get(url, headers=headers)
    print(response)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        books_read = []
        books_want_to_read = []
        books_currently_reading = []

        # Get books read
        read_books_section = soup.find('div', {'class': 'section read'})
        if read_books_section:
            books_read.extend([a.text.strip() for a in read_books_section.find_all('a', {'class': 'bookTitle'})])

        # Get books want to read
        want_to_read_section = soup.find('div', {'class': 'section to-read'})
        if want_to_read_section:
            books_want_to_read.extend([a.text.strip() for a in want_to_read_section.find_all('a', {'class': 'bookTitle'})])

        # Get books currently reading
        currently_reading_section = soup.find('div', {'class': 'section currently-reading'})
        if currently_reading_section:
            books_currently_reading.extend([a.text.strip() for a in currently_reading_section.find_all('a', {'class': 'bookTitle'})])

        return {
            'books_read': books_read,
            'books_want_to_read': books_want_to_read,
            'books_currently_reading': books_currently_reading
        }
    else:
        print("Failed to fetch data from Goodreads")
        return None

# Example usage
username = '115130270-ben-rombaut'
user_books = get_user_books(username)
if user_books:
    print(f"Books Read by {username}:")
    for book in user_books['books_read']:
        print(book)
    
    print(f"\nBooks Want to Read by {username}:")
    for book in user_books['books_want_to_read']:
        print(book)
    
    print(f"\nBooks Currently Reading by {username}:")
    for book in user_books['books_currently_reading']:
        print(book)
```
